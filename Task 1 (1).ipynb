```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['productid']=le.fit_transform(df['productid'])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['score'],bins=5,kde=True)
plt.title('distribution of score')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['helpfulnessnumerator'])
plt.title('Helpfulness Numerator Boxplot')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['helpfulnessnumerator'], y=df['helpfulnessnumerator'])
plt.title('Helpfulness Numerator vs Denominator')
plt.xlabel('Helpfulness Numerator')
plt.ylabel('Helpfulness Denominator')
plt.show() 
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.drop('summary',axis=1).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
